## Importing the Libraries 

```python
import numpy as np
import pandas as pd
```

## Importing the Dataset

```python
dataset= pd.read_csv("mushrooms.csv")
dataset
```

```python
dataset.head()
```

## Data Preprocessing

```python
dataset.info()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
for i in dataset.columns:
    dataset[i]= encoder.fit_transform(dataset[i])
```

```python
dataset
```

```python
dataset.info()
```

```python
X= dataset.iloc[:,1:].values
y= dataset.iloc[:,0].values
```

## Training Set and Data Set ( Splitting the Dataset into Training Set and Test Set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25)
```

## Features Scalling

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
X_train1=X_train
y_train1= y_train
X_test1=X_test
y_test1=y_test
```

## Model ( Logistic Regression)

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train,y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier2= DecisionTreeClassifier()
classifier2.fit(X_train1,y_train1)
```

## Predicting Result and Confusion Metrix and Accuarcy Score with Logistic Regression

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm= confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

## Predicting Result and Confusion Metrix and Accuarcy Score with Decision Tree Classification

```python
y_pred1=classifier2.predict(X_test1)
y_pred1
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm= confusion_matrix(y_test1,y_pred1)
print(cm)
accuracy_score(y_test1,y_pred1)
```
